## Aprendizaje supervisado

En el problema que estamos abordando, queremos predecir el nivel de contaminación del aire y sonora en diferentes áreas de la ciudad de Nueva York. Para lograr esto, necesitamos datos etiquetados que representen el nivel de contaminación en esas áreas. Estos datos etiquetados se utilizarán para entrenar un modelo de machine learning que pueda hacer predicciones sobre la contaminación en áreas nuevas y no vistas.
En este caso, los datos que recopilamos inicialmente incluirían medidas de calidad del aire y contaminación sonora en diferentes áreas de la ciudad, junto con las etiquetas correspondientes que indican el nivel de contaminación en esas áreas. Utilizaremos esta información etiquetada para entrenar nuestro modelo de clasificación supervisada.
Por lo tanto, en resumen, el problema propuesto es un problema de clasificación supervisada, ya que estamos utilizando datos etiquetados para entrenar nuestro modelo y hacer predicciones sobre nuevas instancias.
El objetivo sería que el modelo aprenda de estos datos y sea capaz de hacer predicciones sobre la contaminación en áreas no vistas anteriormente.
Entonces, en lugar de hacer predicciones sobre áreas ya conocidas como altamente contaminadas, el modelo se entrenaría para reconocer patrones en los datos que indiquen niveles de contaminación. Luego, una vez entrenado, se podría utilizar para hacer predicciones sobre áreas nuevas o no observadas en base a las características ambientales que poseen.

Si existen regulaciones ambientales que limitan las emisiones de vehículos en ciertas áreas de la ciudad, la empresa puede utilizar esta información para asegurarse de cumplir con estas regulaciones y evitar posibles sanciones o multas.
