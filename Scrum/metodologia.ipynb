# Metodología SCRUM

## Introducción

## ¿Que es una metodología ágil?

La agilidad se caracteriza por la capacidad de adaptarse de manera flexible a los cambios, buscando maximizar los beneficios. Implica tomar decisiones rápidas y eficientes que conduzcan a resultados satisfactorios.

## ¿Que es SCRUM?

Scrum es un enfoque de desarrollo ágil que se basa en la iteración continua y la colaboración del equipo. Se centra en la entrega temprana y frecuente de partes del producto, en lugar de esperar hasta que todo esté completo. Prioriza la adaptabilidad y la respuesta rápida a medida que se van descubriendo nuevos requisitos y desafíos, en lugar de adherirse a un plan rígido desde el principio.

## ¿Cómo se usa? 

La metodología Scrum implica una colaboración estrecha entre el equipo de desarrollo y el cliente, lo que genera un mayor compromiso y entusiasmo por el proyecto. Al ver el producto crecer de manera incremental, el cliente puede realizar ajustes o cambios según las necesidades del negocio en cualquier momento. Esto no solo facilita la alineación del software con los objetivos empresariales, sino que también fomenta la motivación y el compromiso del equipo. Los profesionales encuentran un entorno favorable para potenciar sus habilidades y contribuir al éxito del proyecto de manera significativa.

## Beneficios de SCRUM

Scrum ofrece una serie de beneficios que incluyen:

- Entrega temprana de valor: El cliente puede comenzar a utilizar funcionalidades importantes del proyecto incluso antes de su finalización total, lo que proporciona beneficios tangibles de manera más rápida.

- Participación activa del cliente: El cliente tiene la oportunidad de establecer expectativas claras y priorizar los requisitos según su valor percibido, lo que garantiza que el producto final se alinee estrechamente con sus necesidades y expectativas.

- Reducción de riesgos: Al trabajar en iteraciones cortas y frecuentes, Scrum permite identificar y abordar los problemas de manera oportuna, lo que ayuda a mitigar los riesgos y a mantener el proyecto en curso.

- Estimaciones más precisas: A través de la planificación de sprints y la revisión continua del progreso, Scrum facilita la predicción de tiempos y la estimación más precisa de la duración de las actividades futuras.

- Mayor productividad: La estructura y los principios de Scrum fomentan la colaboración, la transparencia y la autonomía del equipo, lo que conduce a una mayor productividad y eficiencia en la entrega del proyecto.

# Documentos del SCRUM

## Product backlog

El Product Backlog es una lista dinámica y priorizada de todos los requisitos, funcionalidades y mejoras deseadas para el proyecto. Estos elementos están descritos de manera general y se clasifican según su valor para el negocio. El Product Backlog define qué se va a construir y está constantemente abierto a cambios y ajustes por parte de cualquier miembro del equipo.

## Sprint backlog

El Sprint Backlog es una lista detallada de las tareas que el equipo planea realizar durante el próximo Sprint para implementar los requisitos del Product Backlog.
Si alguna tarea es demasiado grande, se descompone en tareas más pequeñas para una mejor gestión. Las tareas en el Sprint Backlog no están asignadas a miembros específicos del equipo, permitiendo que los miembros del equipo elijan las tareas que desean completar según su capacidad y habilidades.

## Burndown

La gráfica de Burn Down es una herramienta visual utilizada para rastrear el progreso del proyecto durante cada Sprint. Muestra la cantidad de trabajo pendiente en el Backlog del proyecto al inicio de cada Sprint. Al conectar los puntos de los Sprints completados, podemos observar cómo disminuye la cantidad de trabajo a lo largo del tiempo. Idealmente, esta línea seguirá descendiendo hasta alcanzar el eje horizontal, lo que indica que el proyecto se ha completado. La gráfica proporciona una visión clara del avance del equipo y ayuda a identificar posibles desviaciones en el ritmo de trabajo.

## Daily SCRUM

La reunión diaria, también conocida como Daily Scrum o Daily Standup, comienza puntualmente a la hora programada y se lleva a cabo en la misma ubicación todos los días. Todos los participantes se mantienen de pie para fomentar la brevedad y la atención. Durante la reunión, cada miembro del equipo responde tres preguntas:

- ¿Qué avances has logrado desde la última reunión?
- ¿Qué planeas hacer hoy?
- ¿Te has enfrentado a algún obstáculo que necesitas ayuda para resolver?

## Reunión de planificación del Sprint

Al comienzo de cada Sprint, se realiza una "Reunión de Planificación del Sprint", donde el equipo selecciona las tareas que realizará durante el ciclo, con un límite de tiempo de ocho horas. Al final del Sprint, se llevan a cabo dos reuniones: la "Reunión de Revisión del Sprint", donde se muestra el trabajo completado al cliente o stakeholders, y la "Retrospectiva del Sprint", donde el equipo reflexiona sobre el Sprint y busca mejoras para futuras iteraciones.

## Reunión de revisión del Sprint

En la Reunión de Revisión del Sprint, el equipo revisa tanto el trabajo completado como el no completado durante el Sprint. Se presenta el trabajo finalizado a los interesados, pero el trabajo incompleto no se muestra. La reunión tiene un límite de cuatro horas para mantenerla enfocada y productiva.

## Retrospectiva del Sprint

Todos los miembros del equipo comparten sus impresiones y experiencias sobre el Sprint recién finalizado. El objetivo es identificar áreas de mejora en el proceso y promover la mejora continua. La reunión tiene una duración fija de cuatro horas para garantizar que haya suficiente tiempo para discutir y proponer acciones de mejora sin alargarse demasiado.
